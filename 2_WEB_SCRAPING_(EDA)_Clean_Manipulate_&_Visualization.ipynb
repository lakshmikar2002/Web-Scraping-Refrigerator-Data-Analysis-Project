### Whirlpool has the highest number of products and total revenue, showing strong market dominance, while Samsung and LG also contribute significantly to overall sales value.

```python
df.groupby('BRAND').agg(
    no_of_pro = ('BRAND','count'),
    avg_price = ('PRICE','mean'),
    avg_water_l = ("WATER_CAPACITY", 'mean'),
    avg_discount = ('PRICE_OFF','mean'),
    total_per_brand = ("PRICE",'sum')
)
```

### This third table shows the minimum and maximum price for each brand, calculated using Group By and aggregation functions.

```python
df.groupby('BRAND').agg(
    min_pri_brand = ("PRICE","min"),
    max_pri_brand = ("PRICE","max"),
)
```

### used pivote_table anlayse between the two columns.

```python
df.pivot_table(
    index = 'BRAND',
    values = 'PRICE',
    aggfunc = 'sum'
)
```

# VISUALIZATION

## Univariate Analysis
- univariate analysis is a type of data visualization.
- where we visualize only a single variable at a time. this help us to analyze the distribtion of the variable present in the data.

#### Using bar plot, we observe that which company has a greater number of products.

```python
df['BRAND'].value_counts().plot.bar()
```

#### Comparing all company products, which company has more price products. here we use groupby and sum for calculating brandwise.
- the Whirlpool company had a more number of prices, then second is Haier.

```python
df.groupby('BRAND')['PRICE'].sum().plot.bar()
```

## Bivariate Analysis
- Bivariate Analysis is a statistical method used to explore the relationship between two variables. the goal is to undersand whether and how the two variables are related

#### About the Graph:
- Shows average price comparison across brands.
- Some brands have significantly higher average prices (premium brands).
- Some brands offer lower average pricing (budget brands).
- Helps understand brand positioning in terms of pricing strategy.

```python
data = df.groupby('BRAND')['PRICE'].mean()
colors = np.random.rand(len(data), 3)

data.plot.bar(color=colors)
```

#### What It Shows:
- We used to plot,Mostly boxplot is showing outliers in the data, but here undersanding of greater prices product
- The middle line inside each box → Median price.
- The box → Middle 50% of prices (IQR).
- The whiskers → Overall spread of data.
- The dots → Outliers (very high or low prices).

```python
plt.figsize=(10,16)
sns.boxplot(data=df, x= 'BRAND', y='PRICE')
plt.xticks(rotation=90)
plt.show()
```

#### We use histogram plot for understanding of price distribution, where we can observe that around 20k has a greater number of products
- kind = hist → Creates a histogram.
- bins=20 → Divides the price range into 20 intervals.
- It shows how prices are distributed in your dataset.
- Most prices are concentrated in the lower range (around 15k–30k).
- There are few high-price values, creating a right-skewed distribution

```python
df['PRICE'].plot(kind='hist', bins=20, title='Price Distribution')
```

```python
sns.histplot(df['PRICE'], bins=20, kde=True)
```

#### About Graph
-  The chart shows average water capacity for each brand.
- Some brands have higher average capacity (taller bars).
- Some brands have lower average capacity (shorter bars).
- This helps compare which brand generally offers larger capacity products.

```python
avg_price = df.groupby('BRAND')['WATER_CAPACITY'].mean()

plt.bar(avg_price.index, avg_price.values)
plt.xticks(rotation=90)
plt.show()
```

#### About Graph
- Shows average water capacity per brand.
- Taller bars indicate brands with higher average capacity.
- Black vertical lines show variation in data.
- Useful to compare brands and understand capacity consistency.

```python
sns.barplot(data=df, x='BRAND', y='WATER_CAPACITY')
plt.xticks(rotation=90);
```

## Multivariate Analysis
- Multivariate Analysis is a statistical technique used to analyze datasets with two or more variables. It help uncover relationships, reduce complexity and interpret underlaying structures in data,

#### About Graph
- Shows the relationship between capacity and price.
- There is a positive correlation — as capacity increases, price generally increases.
- Higher-capacity products (500L+) tend to have higher prices.
- Some variation exists, but overall trend is upward.

```python
plt.scatter(df['WATER_CAPACITY'], df['PRICE'])
plt.xlabel('Capacity (L)')
plt.ylabel('Price')
plt.show()
```

#### About Graph
- Shows the relationship between capacity and price for each brand.
- There is a positive trend — higher capacity generally means higher price.
- Different colors help compare how brands price their products.
- Some brands have premium pricing even at similar capacities.

```python
sns.scatterplot(data=df, x='WATER_CAPACITY', y='PRICE', hue='BRAND')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### About the Graph:
- Shows the distribution of product ratings (stars).
